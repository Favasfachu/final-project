```python
data['Cars Names']=data['Cars Names'].astype('string')
data['Engines'] = data['Engines'].astype('string')
data['Fuel Types']=data['Fuel Types'].astype('category')

```

```python
data['Total Speed']=data['Total Speed'].str.replace('km-h','',regex=False)
data['HorsePower']=data['HorsePower'].str.replace('hp','',regex=False)
data['Torque']=data['Torque'].str.replace('Nm','',regex=False)
```

```python
data['Performance(0 - 100 )KM-H'] = data['Performance(0 - 100 )KM-H'].astype(str).str.extractall(r'(\d+\.?\d*)') \
    .groupby(level=0)[0].apply(lambda x: x.astype(float).max())
data['HorsePower'] = (
    data['HorsePower'].astype(str)
    .str.extractall(r'(\d+\.?\d*)')
    .groupby(level=0)[0]
    .apply(lambda x: x.astype(float).max())
)
data['Torque'] = (
    data['Torque'].astype(str)
    .str.extractall(r'(\d+\.?\d*)')
    .groupby(level=0)[0]
    .apply(lambda x: x.astype(float).max())
)
```

```python
data['Cars Prices'] = (
    data['Cars Prices'].astype(str)
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('–', '-', regex=False)
    .str.split('-')
    .apply(lambda x: max(
        [y.strip() for y in x if any(c.isdigit() for c in y)],
        default=None
    ))
)

```

```python
data=data.iloc[:-5]
```

```python
data = data[~data['CC-Battery Capacity'].str.contains('kWh|electric', case=False, na=False)]
data = data[~data['CC-Battery Capacity'].str.contains('(V6)|(V8)|(I4)', case=False, na=False)]
```

```python
data['CC-Battery Capacity']=data['CC-Battery Capacity'].str.replace('kWh','',regex=False)
data['CC-Battery Capacity']=data['CC-Battery Capacity'].str.replace('cc','',regex=False)
data['CC-Battery Capacity'] = data['CC-Battery Capacity'].astype(str).str.replace(',', '', regex=False)
```

```python
data['CC-Battery Capacity'] = data['CC-Battery Capacity'].astype(str).str.split('-').str[-1].str.strip()
data['Engines'] = data['Engines'].astype(str).str.split('-').str[-1].str.strip()
data['Fuel Types'] = data['Fuel Types'].astype(str).str.split('-').str[-1].str.strip()
```

```python
data['Cars Prices']=data['Cars Prices'].astype(float)
data['Total Speed']=data['Total Speed'].astype(float)
```

```python
data['Fuel Types']=data['Fuel Types'].astype('category')
data['Engines']=data['Engines'].astype('string')
```

```python
# Convert to numeric safely; non-numeric values become NaN
data['CC-Battery Capacity'] = pd.to_numeric(data['CC-Battery Capacity'], errors='coerce')

#Convert to nullable integer
data['CC-Battery Capacity'] = data['CC-Battery Capacity'].astype('Int64')
```

```python
# DERIVED COLUMNS
bins = [0, 50000, 100000, 1000000, float('inf')]
labels = ['Budget','Mid-Range', 'Premium', 'Luxury']

data['Price_Category'] = pd.cut(data['Cars Prices'], bins=bins, labels=labels)
```

```python
bins = [0, 150, 400, 1000, float('inf')]
labels = ['Low Power', 'Medium Power', 'High Power', 'Super car']

data['Power_Level'] = pd.cut(data['HorsePower'], bins=bins, labels=labels, right=False)
```

```python
data['Value_for_Money'] = data['HorsePower'] / data['Cars Prices'] * 1000 #higher numbers indicates most value for monbey cars
```
